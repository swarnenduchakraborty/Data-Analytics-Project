# Data Cleaning

clearing the data

```python
df.head(1)
```

```python
# drop Cabin data because it have move than 50 % nan values
df.drop(columns=['Cabin'],inplace = True)
```

```python
df.info()
```

```python
df.head()
```

```python
# now clear others nan values
df.isnull().sum()
```

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
# if there is less diff in mean and median then fill data with mean()
# if there is high diff in mean and median then fill data with median()
# now fill
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
# As you can see nan will be fixed
```

```python
df['Embarked'].mode()
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

Now as you see your data has been clear now!!!

# EXPLORATORY DATA ANALYSIS

```python
# TELL HOW MANY SURVIVED in %

df['Survived'].value_counts()/df.shape[0]*(100)
```

```python
# Tell how many people travels in each class

df['Pclass'].value_counts()/df.shape[0]*(100)
```

```python
# Tell how many people are male or female

df['Sex'].value_counts()/df.shape[0]*(100)
```

```python
# find data where age is max and Fare is max

df['Age'].max()
```

```python
df['Fare'].max()
```

```python
df[(df['Age']==max(df['Age']))  | (df['Fare']==max(df['Fare']))]
```

```python
# find data where age is min and Fare is min

df['Fare'].min()
```

```python
df['Age'].min()
```

```python
df[(df['Age']==min(df['Age']))  | (df['Fare']==min(df['Fare']))]
```

```python
df.set_index("PassengerId", inplace=True)
```

```python
df
```

```python
# find where age is greater than 35 and fare is greater than 100

df[(df['Age']>35) & (df['Fare']>100)]
```

```python
# Tell the mean min mac sum count od sex and fair using aggregation

df.groupby(['Sex'])['Fare'].agg(['mean','min','max','sum','count'])
```

```python
# Tell that how many male and female survived in titanic

df.groupby(['Pclass','Sex'])['Survived'].value_counts()
```
